```python
import time
import numpy as np
import time
from keras.preprocessing.image import ImageDataGenerator, array_to_img,img_to_array,load_img
from keras.applications.inception_resnet_v2 import preprocess_input,decode_predictions
from keras.applications.inception_resnet_v2 import InceptionResNetV2
```

```python
img_inceptionv2_pth = '/Users/shreyassetlurarun/Downloads/dogs-vs-cats/test1/28.jpg'
img_inceptionv2 = load_img(img_inceptionv2_pth,target_size=(299,299))
img_inceptionv2
```

```python
img_inceptionv2 = img_to_array(img_inceptionv2)
print(img_inceptionv2.shape)
```

```python
img_inceptionv2 = np.expand_dims(img_inceptionv2,axis = 0)
print(img_inceptionv2.shape)
```

```python
img_inceptionv2 = preprocess_input(img_inceptionv2)
print(img_inceptionv2.shape)
```

```python
tic = time.perf_counter()
model_inceptionv2 = InceptionResNetV2()
toc = time.perf_counter()
print(f"Time for loading the model is {toc - tic:0.4f} seconds")
```

```python
#Prediction
tic = time.perf_counter()
pred_inceptionv2 = model_inceptionv2.predict(img_inceptionv2)
n = 10
top_n_inceptionv2 = decode_predictions(pred_inceptionv2,top = n)
for a in top_n_inceptionv2[0]:
    print(a)
toc = time.perf_counter()
print(f"Time for prediction is {toc - tic:0.4f} seconds")
```
