```python
import numpy as np
import time
from keras.preprocessing.image import ImageDataGenerator, array_to_img,img_to_array,load_img
from keras.applications.mobilenet_v2 import preprocess_input,decode_predictions
from keras.applications.mobilenet_v2 import MobileNetV2
```

```python
img_mobilenetv2_pth = '/Users/shreyassetlurarun/Downloads/dogs-vs-cats/test1/28.jpg'
img_mobilenetv2 = load_img(img_mobilenetv2_pth,target_size=(224,224))
img_mobilenetv2
```

```python
img_mobilenetv2 = img_to_array(img_mobilenetv2)
print(img_mobilenetv2.shape)
```

```python
img_mobilenetv2 = np.expand_dims(img_mobilenetv2,axis = 0)
print(img_mobilenetv2.shape)
```

```python
img_mobilenetv2 = preprocess_input(img_mobilenetv2)
print(img_mobilenetv2.shape)
```

```python
tic = time.perf_counter()
model_mobilenetv2 = MobileNetV2()
toc = time.perf_counter()
print(f"Time for loading the model is {toc - tic:0.4f} seconds")
```

```python
#Prediction

tic = time.perf_counter()
pred_mobilenetv2 = model_mobilenetv2.predict(img_mobilenetv2)
n = 10
top_n_mobilenetv2 = decode_predictions(pred_mobilenetv2,top = n)
for a in top_n_mobilenetv2[0]:
    print(a)
toc = time.perf_counter()
print(f"Time for prediction is {toc - tic:0.4f} seconds")
```
